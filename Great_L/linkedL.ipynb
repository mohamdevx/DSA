```python
# Node class for each element in the singly linked list
class Node:
    def __init__(self, data):
        self.data = data  # Value of the node
        self.next = None  # Pointer to the next node

# Singly Linked List (SLL) class with insertion and traversal methods
class Sll:
    def __init__(self):
        self.head = None  # Start of the linked list

    # Traverse the linked list and print each node's data
    def traversal(self):
        if self.head is None:
            print("Linked list is empty")
            return
        else:
            a = self.head
            while a is not None:
                print(a.data, end=" ")
                a = a.next

    # Insert a node at the beginning of the list
    def insert_at_beginning(self, data):
        nb = Node(data)
        nb.next = self.head
        self.head = nb

    # Insert a node at the end of the list
    def insert_at_end(self, data):
        ne = Node(data)
        a = self.head
        while a.next is not None:
            a = a.next
        a.next = ne

    # Insert a node at a specified position (1-based index)
    def insert_at_specified_node(self, position, data):
        nib = Node(data)
        a = self.head
        for i in range(1, position - 1):
            a = a.next
        nib.next = a.next
        a.next = nib
        
    
    def deletion_at_beginning(self):   
        a=self.head
        self.head=a.next
        a.next=None
        
        
    def deletion_at_end(self):
        prev=self.head
        a=self.head.next
        while a.next is not None:
            a=a.next
            prev=prev.next
        prev.next=None
                
        
    def deletion_at_particular_node(self,position):
        prev=self.head
        a=self.head.next
        for i in range(1,position-1):
            a=a.next
            prev=prev.next
        prev.next=a.next
        a.next=None
            
        
        

# Creating initial nodes and linking them
n1 = Node(5)
sll = Sll()
sll.head = n1
n2 = Node(10)
n1.next = n2
n3 = Node(15)
n2.next = n3
n4 = Node(20)
n3.next = n4
sll.head = n1

# Traversal before any insertion
sll.traversal()

# Insert at beginning
print("\ninsert at beginning")
sll.insert_at_beginning(2)
sll.traversal()

# Insert at end
sll.insert_at_end(25)
print("\ninsert at the end")
sll.traversal()

# Insert at specified position (3rd position)
sll.insert_at_specified_node(3, 8)
print("\nat specified position")
sll.traversal()

sll.deletion_at_beginning()
print("\ndeletion at beginning")
sll.traversal()

sll.deletion_at_end()
print("\ndeletion at end")
sll.traversal()

sll.deletion_at_particular_node(3)
print("\ndelete at particular node")
sll.traversal()
```
