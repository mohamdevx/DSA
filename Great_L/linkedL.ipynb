```python
# Node class for each element in the singly linked list
class Node:
    def __init__(self, data):
        self.data = data  # Value of the node
        self.next = None  # Pointer to the next node

# Singly Linked List (SLL) class with insertion and traversal methods
class Sll:
    def __init__(self):
        self.head = None  # Start of the linked list

    # Traverse the linked list and print each node's data
    def traversal(self):
        if self.head is None:
            print("Linked list is empty")
            return
        else:
            a = self.head
            while a is not None:
                print(a.data, end=" ")
                a = a.next

    # Insert a node at the beginning of the list
    def insert_at_beginning(self, data):
        nb = Node(data)
        nb.next = self.head
        self.head = nb

    # Insert a node at the end of the list
    def insert_at_end(self, data):
        ne = Node(data)
        a = self.head
        while a.next is not None:
            a = a.next
        a.next = ne

    # Insert a node at a specified position (1-based index)
    def insert_at_specified_node(self, position, data):
        nib = Node(data)
        a = self.head
        for i in range(1, position - 1):
            a = a.next
        nib.next = a.next
        a.next = nib
        
    
    def deletion_at_beginning(self):   
        a=self.head
        self.head=a.next
        a.next=None
        
        
    def deletion_at_end(self):
        prev=self.head
        a=self.head.next
        while a.next is not None:
            a=a.next
            prev=prev.next
        prev.next=None
                
        
    def deletion_at_particular_node(self,position):
        prev=self.head
        a=self.head.next
        for i in range(1,position-1):
            a=a.next
            prev=prev.next
        prev.next=a.next
        a.next=None
            
        
        

# Creating initial nodes and linking them
n1 = Node(5)
sll = Sll()
sll.head = n1
n2 = Node(10)
n1.next = n2
n3 = Node(15)
n2.next = n3
n4 = Node(20)
n3.next = n4
sll.head = n1

# Traversal before any insertion
sll.traversal()

# Insert at beginning
print("\ninsert at beginning")
sll.insert_at_beginning(2)
sll.traversal()

# Insert at end
sll.insert_at_end(25)
print("\ninsert at the end")
sll.traversal()

# Insert at specified position (3rd position)
sll.insert_at_specified_node(3, 8)
print("\nat specified position")
sll.traversal()

sll.deletion_at_beginning()
print("\ndeletion at beginning")
sll.traversal()

sll.deletion_at_end()
print("\ndeletion at end")
sll.traversal()

sll.deletion_at_particular_node(3)
print("\ndelete at particular node")
sll.traversal()
```

```python
# Node class for doubly linked list
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None


# Doubly linked list class
class Dll:
    def __init__(self):
        self.head = None

    # Forward traversal
    def forward_traversal(self):
        if self.head is None:
            print("Doubly linked list is empty")
            return
        a = self.head
        while a is not None:
            print(a.data, end=" ")
            a = a.next

    # Backward traversal
    def backward_traversal(self):
        if self.head is None:
            print("Doubly linked list is empty")
            return
        a = self.head
        while a.next is not None:
            a = a.next  # Go to the last node
        while a is not None:
            print(a.data, end=" ")
            a = a.prev  # Traverse back

    # Insertion at beginning
    def insert_at_beggining(self, data):
        ns = Node(data)
        a = self.head
        a.prev = ns
        ns.next = a
        self.head = ns

    # Insertion at end
    def insertion_at_end(self, data):
        ne = Node(data)
        a = self.head
        while a.next is not None:
            a = a.next
        a.next = ne
        ne.prev = a

    # Insertion at specified position
    def insertion_at_specified_node(self, position, data):
        nib = Node(data)
        a = self.head
        for i in range(1, position - 1):
            a = a.next
        nib.prev = a
        nib.next = a.next
        a.next.prev = nib
        a.next = nib

    # Deletion at beginning
    def deletion_at_beginning(self):
        a = self.head
        self.head = a.next
        a.next = None
        self.head.prev = None

    # Deletion at end
    def deletion_at_end(self):
        prev = self.head
        a = self.head.next
        while a.next is not None:
            a = a.next
            prev = prev.next
        prev.next = None
        a.prev = None

    # Deletion at particular node
    def deletion_at_particular_node(self, position):
        a = self.head.next
        b = self.head
        for i in range(1, position - 1):
            a = a.next
            b = b.next
        b.next = a.next
        a.next.prev = b
        a.next = None
        a.prev = None


# Create initial nodes and link them
n1 = Node(5)
dll = Dll()
dll.head = n1

n2 = Node(10)
n1.next = n2
n2.prev = n1

n3 = Node(15)
n2.next = n3
n3.prev = n2

n4 = Node(20)
n3.next = n4
n4.prev = n3

# Traversals
print("Forward traversal:")
dll.forward_traversal()

print("\nBackward traversal:")
dll.backward_traversal()

# Insert at beginning
dll.insert_at_beggining(2)
print("\n\nAfter inserting 2 at the beginning:")
dll.forward_traversal()

print("\nBackward traversal:")
dll.backward_traversal()

# Insert at end
dll.insertion_at_end(25)
print("\n\nAfter inserting 25 at the end:")
dll.forward_traversal()

print("\nBackward traversal:")
dll.backward_traversal()

# Insert at position 3
dll.insertion_at_specified_node(3, 7)
print("\n\nAfter inserting 7 at position 3:")
dll.forward_traversal()

print("\nBackward traversal:")
dll.backward_traversal()

# Delete at beginning
dll.deletion_at_beginning()
print("\n\nAfter deleting at the beginning:")
dll.forward_traversal()

# Delete at end
dll.deletion_at_end()
print("\n\nAfter deleting at the end:")
dll.forward_traversal()

print("\nBackward traversal:")
dll.backward_traversal()

# Delete at position 3
dll.deletion_at_particular_node(3)
print("\n\nAfter deletion at particular node (position 3):")
dll.forward_traversal()
```
